## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions:
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

## Q1. Based on the trends for the share of CVM claims as a percentage of total claims over the years 2016 through 2018, what are some business insights you can gather? What are some additional analyses you could do based on these trends?

For this question, I looked at the top 100 HCPCS/CPT codes and grouped them into CVM and non-CVM claims. I used the category column from the top 100 code table and treated the Cardiac category as CVM. Then I found the percentage of CVM claims by year from 2016 to 2018 and made a 100% stacked bar chart.

```python
import matplotlib.pyplot as plt

# copy the data
q1_codes = hcpcs_code_100_df.copy()
q1_data = medicare_df.copy()

# make sure both hcpcs code columns are strings
q1_codes["hcpcs_code"] = q1_codes["hcpcs_code"].astype(str)
q1_data["hcpcs_code"] = q1_data["hcpcs_code"].astype(str)

# merge the main data with the top 100 code table
q1_merged = pd.merge(q1_data, q1_codes[["hcpcs_code", "category"]], on="hcpcs_code", how="inner")

# label cardiac claims as CVM
q1_merged["claim_type"] = q1_merged["category"].apply(lambda x: "CVM" if x == "Cardiac" else "Non-CVM")

# count claims by year and type
q1_summary = q1_merged.groupby(["claim_year", "claim_type"]).size().reset_index(name="count")

# reshape the table
q1_table = q1_summary.pivot(index="claim_year", columns="claim_type", values="count").fillna(0)

# turn counts into percentages
q1_percent = q1_table.div(q1_table.sum(axis=1), axis=0) * 100

print("Q1 claim counts by year")
display(q1_table)

print("Q1 percentage share by year")
display(q1_percent.round(2))

# chart
q1_percent[["CVM", "Non-CVM"]].plot(kind="bar", stacked=True, figsize=(8,5))

plt.title("Share of CVM vs Non-CVM Claims by Year")
plt.xlabel("Claim Year")
plt.ylabel("Percent of Claims")
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.legend(title="Claim Type")
plt.tight_layout()
plt.show()
```

### Q1. Business Insights

From 2016 to 2018, the share of CVM claims increased slightly from 13.08% to 14.16%. Even though the increase is not huge, it still shows that CVM claims were making up a slightly bigger part of the total claims over time. This could mean that cardiovascular-related services were becoming a little more common in this dataset during these years.

This trend could be useful for a pharmaceutical company because it suggests that the CVM area may be growing in importance compared to other claim types. If this trend continued, the company might want to spend more time studying this area, especially to see which procedures, providers, or patient groups are driving the change.

Some additional analyses that could be helpful would be looking at which specific CVM HCPCS/CPT codes increased the most, which providers submitted the most CVM claims, and whether certain patient age groups were more connected to the increase.

## Q2. Evaluate the HCP behavior in context of claim volume from 2016-2018. How many HCPs are submitting 1 CVM claim; how many HCPs are associated with more than 10 claims, etc.? Once you perform this analysis, explain how this trend can influence the sales force deployment. That is, how would you segment the HCPs and how would you allocate In-Person (sales force) vs Non-Personal Promotions (NPP, i.e. Emails, Social Media, Digital etc.) efforts?

For this question, I looked at provider behavior based on how many CVM claims they submitted each year from 2016 to 2018. I grouped providers into four segments based on claim count: 1 claim, 2 to 4 claims, 5 to 9 claims, and 10 or more claims. Then I made a stacked bar chart to compare the number of providers in each group by year.

```python
import matplotlib.pyplot as plt

# keep only cardiac claims from the top 100 code list
q2_codes = hcpcs_code_100_df.copy()
q2_data = medicare_df.copy()

q2_codes["hcpcs_code"] = q2_codes["hcpcs_code"].astype(str)
q2_data["hcpcs_code"] = q2_data["hcpcs_code"].astype(str)

q2_merged = pd.merge(q2_data, q2_codes[["hcpcs_code", "category"]], on="hcpcs_code", how="inner")

q2_cvm = q2_merged[q2_merged["category"] == "Cardiac"].copy()

# remove rows where npi_id is missing since q2 is about providers
q2_cvm = q2_cvm[q2_cvm["npi_id"].notna()].copy()

# count cvm claims per provider per year
q2_provider_counts = q2_cvm.groupby(["claim_year", "npi_id"]).size().reset_index(name="claim_count")

# make the claim volume buckets
def segment_hcp(count):
    if count == 1:
        return "1 claim"
    elif 2 <= count <= 4:
        return "2-4 claims"
    elif 5 <= count <= 9:
        return "5-9 claims"
    else:
        return "10+ claims"

q2_provider_counts["segment"] = q2_provider_counts["claim_count"].apply(segment_hcp)

# count how many providers fall into each segment each year
q2_summary = q2_provider_counts.groupby(["claim_year", "segment"]).size().reset_index(name="hcp_count")

# reshape into a table
q2_table = q2_summary.pivot(index="claim_year", columns="segment", values="hcp_count").fillna(0)

# put columns in a cleaner order
q2_table = q2_table[["1 claim", "2-4 claims", "5-9 claims", "10+ claims"]]

print("Q2 HCP counts by segment and year")
display(q2_table)

# chart
q2_table.plot(kind="bar", stacked=True, figsize=(8,5))

plt.title("HCP CVM Claim Volume Segments by Year")
plt.xlabel("Claim Year")
plt.ylabel("Number of HCPs")
plt.xticks(rotation=0)
plt.legend(title="HCP Segment")
plt.tight_layout()
plt.show()
```

### Q2. Business Insights

The results show that most HCPs were in the 1 claim segment in every year, and that group also grew from 2,140 in 2016 to 2,724 in 2018. The 2 to 4 claims group also increased over time, while the 5 to 9 claims and 10+ claims groups stayed much smaller. This suggests that most providers in the dataset had only limited CVM claim activity, while a much smaller number of providers were responsible for higher claim volumes.

This could be useful for sales force planning. Providers in the 10+ claims and 5 to 9 claims groups may be the best fit for more direct sales force attention since they already show stronger CVM activity. The 2 to 4 claims group could be a middle group that may benefit from a mix of in-person outreach and non-personal promotion. The 1 claim group is the largest, so it may make more sense to reach many of those providers through lower-cost non-personal promotion, such as email, digital marketing, or other awareness campaigns.

Overall, this trend suggests that the company should not treat all HCPs the same. It would make more sense to segment them by claim activity and match the type of outreach to the size of each group.**bold text**

## Q3. Evaluate the Patient Age demographics in the context of claim volume from 2016-2018. Bucket the patients into groups based on their age and explain the trends. How would you position the Marketing Budgets and the Promotions with respect to the changing landscape of the CVM claims and the respective patient segments?

For this question, I looked at CVM claims by patient age group from 2016 to 2018. I grouped patients into four age buckets: 18 to 59, 60 to 69, 70 to 79, and 80+. Then I counted the number of CVM claims in each age group by year, made a stacked bar chart, and also calculated the year-over-year percent change for each age group.

```python
import numpy as np
import matplotlib.pyplot as plt

# copy data
q3_codes = hcpcs_code_100_df.copy()
q3_data = medicare_df.copy()

# make sure hcpcs code matches in both tables
q3_codes["hcpcs_code"] = q3_codes["hcpcs_code"].astype(str)
q3_data["hcpcs_code"] = q3_data["hcpcs_code"].astype(str)

# merge with code category table
q3_merged = pd.merge(q3_data, q3_codes[["hcpcs_code", "category"]], on="hcpcs_code", how="inner")

# keep only cardiac claims
q3_cvm = q3_merged[q3_merged["category"] == "Cardiac"].copy()

# keep only rows with age present
q3_cvm = q3_cvm[q3_cvm["patient_age"].notna()].copy()

# make age buckets
def age_group(age):
    if 18 <= age <= 59:
        return "18-59"
    elif 60 <= age <= 69:
        return "60-69"
    elif 70 <= age <= 79:
        return "70-79"
    else:
        return "80+"

q3_cvm["age_group"] = q3_cvm["patient_age"].apply(age_group)

# count claims by year and age group
q3_summary = q3_cvm.groupby(["claim_year", "age_group"]).size().reset_index(name="claim_count")

# pivot table
q3_table = q3_summary.pivot(index="claim_year", columns="age_group", values="claim_count").fillna(0)

# put columns in order
q3_table = q3_table[["18-59", "60-69", "70-79", "80+"]]

print("Q3 CVM claim counts by age group and year")
display(q3_table)

# stacked bar chart
q3_table.plot(kind="bar", stacked=True, figsize=(8,5))

plt.title("CVM Claims by Patient Age Group and Year")
plt.xlabel("Claim Year")
plt.ylabel("Number of Claims")
plt.xticks(rotation=0)
plt.legend(title="Age Group")
plt.tight_layout()
plt.show()

# year-over-year percent change
q3_yoy = q3_table.pct_change() * 100

print("Q3 Year-over-Year Percent Change")
display(q3_yoy.round(2))
```

### Q3. Business Insights

The results show that the 70 to 79 and 80+ age groups had the highest number of CVM claims from 2016 to 2018. All age groups increased over time, but the largest growth happened in the older patient groups. The 80+ group had the biggest year-over-year increase from 2017 to 2018, and the 70 to 79 group also grew a lot during the same time period.

This suggests that older patients are a very important part of the CVM claim landscape. Because of that, the company may want to spend more of its marketing budget on campaigns, education, and outreach that are more relevant to older age groups. At the same time, the younger age groups should not be ignored, but they may not need as much focus if the goal is to target the largest and fastest-growing segments.

Overall, this trend suggests that promotions and marketing strategy should be more focused on older populations, especially patients age 70 and above, since they make up a large share of the claims and are also showing strong growth over time.
